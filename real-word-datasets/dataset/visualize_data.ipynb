```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_array_from_csv(file_name):
  with open(file="./real-word-datasets/dataset_csv/" + file_name, mode='r') as file:
    df = pd.read_csv(file)    
  array = pd.DataFrame.to_numpy(df)
  return(array)
```

```python
list_names = ["ar10p.csv","cll_sub.csv","orl10p.csv","pie10p.csv","pix10p.csv","tox.csv"]
data = {}
for file_name in list_names:
    data[str(file_name)] = get_array_from_csv(file_name)
```

```python
def plot_datasets_as_images(data):
    plt.figure(figsize=(15, 10))
    for i, (name, array) in enumerate(data.items()):
        plt.subplot(2, 3, i + 1)  # Create a subplot grid (2 rows, 3 columns)
        plt.imshow(array, cmap='gray', aspect='auto')
        plt.title(name.replace(".csv", ""))
        plt.axis('off')  # Turn off the axis for better visualization
    plt.tight_layout()
    plt.show()
```

```python
plot_datasets_as_images(data)
```
