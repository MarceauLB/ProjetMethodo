```python
# usefuls packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pyHSICLasso import HSICLasso
```

#### results features selections data1 hsic lasso

```python
rep = 30
res_matrix1 = np.zeros((10,rep))

p = 256
for i in range(0,10):
  n = (i + 1)*25
  print(n)
  res_repetitions = np.zeros((1,rep))
  for j in range(0,rep):
    np.random.seed(i * rep + j)  # Ensure different seeds
    X = np.random.normal(size=(n,p))
    eps = np.random.normal(size=n)
    y = -2*np.sin(2*X[:,4]) + X[:,5]**2 + X[:,6] + np.exp(-X[:,7]) + eps

    hsic_lasso = HSICLasso()
    hsic_lasso.input(X,y)
    hsic_lasso.regression(num_feat=4,covars_kernel='Gaussian')

    features_index = hsic_lasso.get_index()
    res_repetitions[0,j] = len(set(features_index).intersection(set([4,5,6,7])))/4*100
  res_matrix1[i] = res_repetitions
```

```python
res_matrix1
```

```python
rmean1 = res_matrix1.mean(axis=1)
rmean1
```

```python
sample_size = [(i+1)*25 for i in range(0,10)]
plt.plot(sample_size, rmean1/100, color='red', linestyle='-', marker='o')
plt.ylim(-0.1, 1.1)
plt.show()
```

### HSIC Data 2

```python
rep = 30
res_matrix2 = np.zeros((10,rep))

p = 1000
for i in range(0,10):
  n = (i + 1)*25
  print(n)
  res_repetitions = np.zeros((1,rep))
  for j in range(0,rep):
    np.random.seed(i * rep + j)  # Ensure different seeds
    X = np.random.normal(size=(n,p))
    eps = np.random.normal(size=n)
    y = X[:,3]*np.exp(2*X[:,4]) + X[:,5]**2 + eps

    hsic_lasso = HSICLasso()
    hsic_lasso.input(X,y)
    hsic_lasso.regression(num_feat=4,covars_kernel='Gaussian')

    features_index = hsic_lasso.get_index()
    res_repetitions[0,j] = len(set(features_index).intersection(set([3,4,5])))/3*100
  res_matrix2[i] = res_repetitions
```

```python
res_matrix2
```

```python
rmean2 = res_matrix2.mean(axis=1)
rmean2
```

```python
sample_size = [(i+1)*25 for i in range(0,10)]
plt.plot(sample_size, rmean2/100, color='red', linestyle='-', marker='o')
plt.ylim(-0.1, 1.1)
plt.show()
```
